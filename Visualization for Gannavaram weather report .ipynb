### Line Plot

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('10years GannavaramData.csv')

# Rename the relevant columns for clarity
data.columns = ['Date', 'Location', 'Unknown1', 'Max Temp', 'Unknown2', 'Min Temp', 'Unknown3', 'RH% 830', 'RH% 1730', 'Unknown4', 'Rainfall']

# Convert 'Date' to numeric for plotting and temperature columns to floats
data['Date'] = pd.to_numeric(data['Date'], errors='coerce')
data['Max Temp'] = pd.to_numeric(data['Max Temp'], errors='coerce')
data['Min Temp'] = pd.to_numeric(data['Min Temp'], errors='coerce')

# Drop rows where the 'Date' is NaN (these are probably the rows without actual data)
data.dropna(subset=['Date'], inplace=True)

# Plot for Max Temp
plt.figure(figsize=(10,6))
plt.plot(data['Date'], data['Max Temp'], label='Max Temp', color='coral', marker='o')

plt.title('Maximum Temperature Over Time - Gannavaram')
plt.xlabel('Date (November 2013)')
plt.ylabel('Max Temperature (C)')
plt.legend()
plt.grid(True)
plt.show()

# Plot for Min Temp
plt.figure(figsize=(10,6))
plt.plot(data['Date'], data['Min Temp'], label='Min Temp', color='gray', marker='o')

plt.title('Minimum Temperature Over Time - Gannavaram')
plt.xlabel('Date (November 2013)')
plt.ylabel('Min Temperature (C)')
plt.legend()
plt.grid(True)
plt.show()
```

### Heat Map

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV file (adjust file path as necessary)
file_path = '10years GannavaramData.csv'
data = pd.read_csv(file_path)

# Print the columns to understand their structure
print(data.columns)

# Rename columns based on the actual structure
data.columns = ["Date", "Station", "Column3", "Max Temp", "Column5", "Min Temp", "Column7", "RH% 830", "RH% 1730", "Column10", "Rainfall"]

# Drop unnecessary columns (Column3, Column5, Column7, Column10)
data = data.drop(columns=["Column3", "Column5", "Column7", "Column10"])

# Convert relevant columns to numeric values
data["Max Temp"] = pd.to_numeric(data["Max Temp"], errors='coerce')
data["Min Temp"] = pd.to_numeric(data["Min Temp"], errors='coerce')
data["RH% 830"] = pd.to_numeric(data["RH% 830"], errors='coerce')
data["RH% 1730"] = pd.to_numeric(data["RH% 1730"], errors='coerce')
data["Rainfall"] = pd.to_numeric(data["Rainfall"], errors='coerce')

# Remove any rows where 'Date' is not numeric
data = data[data["Date"] != "Date"]
data["Date"] = pd.to_numeric(data["Date"], errors='coerce')

# Select the relevant columns for the heatmap
heatmap_data = data[["Max Temp", "Min Temp", "RH% 830", "RH% 1730", "Rainfall"]]

# Compute the correlation matrix
corr = heatmap_data.corr()

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)

# Add labels and title
plt.title('Heatmap of Weather Data Variables Correlation')
plt.show()
```

```python
import pandas as pd
import squarify
import matplotlib.pyplot as plt

# Load the CSV file (adjust file path as necessary)
file_path = '10years GannavaramData.csv'
data = pd.read_csv(file_path)

# Print the columns to understand their structure
print(data.columns)

# Rename columns based on the actual structure
data.columns = ["Date", "Station", "Column3", "Max Temp", "Column5", "Min Temp", "Column7", "RH% 830", "RH% 1730", "Column10", "Rainfall"]

# Drop unnecessary columns (Column3, Column5, Column7, Column10)
data = data.drop(columns=["Column3", "Column5", "Column7", "Column10"])

# Convert relevant columns to numeric values
data["Max Temp"] = pd.to_numeric(data["Max Temp"], errors='coerce')
data["Min Temp"] = pd.to_numeric(data["Min Temp"], errors='coerce')
data["Rainfall"] = pd.to_numeric(data["Rainfall"], errors='coerce')
data["RH% 830"] = pd.to_numeric(data["RH% 830"], errors='coerce')
data["RH% 1730"] = pd.to_numeric(data["RH% 1730"], errors='coerce')

# Remove any rows where 'Date' is not numeric
data = data[data["Date"] != "Date"]
data["Date"] = pd.to_numeric(data["Date"], errors='coerce')

# Calculate the total values for the treemap
total_values = data[["Max Temp", "Min Temp", "Rainfall", "RH% 830", "RH% 1730"]].sum()

# Create labels and sizes for the treemap
labels = total_values.index
sizes = total_values.values

# Plot the treemap
plt.figure(figsize=(12, 8))
squarify.plot(sizes=sizes, label=labels, alpha=0.7, color=['orangered', 'lightgreen', 'blue', 'gray', 'blueviolet'])

# Add title
plt.title('Treemap of Total Weather Data Variables')

# Remove axes for clarity
plt.axis('off')

# Show the plot
plt.show()
```

### Violin Plot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('10years GannavaramData.csv')
# Print the columns to understand their structure
print(data.columns)

# Rename columns based on the actual structure
data.columns = ["Date", "Station", "Column3", "Max Temp", "Column5", "Min Temp", "Column7", "RH% 830", "RH% 1730", "Column10", "Rainfall"]

# Drop unnecessary columns (Column3, Column5, Column7, Column10)
data = data.drop(columns=["Column3", "Column5", "Column7", "Column10"])

# Convert relevant columns to numeric values
data["Max Temp"] = pd.to_numeric(data["Max Temp"], errors='coerce')
data["Min Temp"] = pd.to_numeric(data["Min Temp"], errors='coerce')
data["Rainfall"] = pd.to_numeric(data["Rainfall"], errors='coerce')

# Remove any rows where 'Date' is not numeric
data = data[data["Date"] != "Date"]
data["Date"] = pd.to_numeric(data["Date"], errors='coerce')

# Prepare the data for the violin plot
melted_data = data.melt(value_vars=["Max Temp", "Min Temp", "Rainfall"], 
                         var_name="Variable", value_name="Value")

# Create the violin plot
plt.figure(figsize=(12, 8))
sns.violinplot(x='Variable', y='Value', data=melted_data,hue='Variable',inner='quartile', palette='pastel')

# Add title and labels
plt.title('Violin Plot of Max Temp, Min Temp, and Rainfall')
plt.xlabel('Weather Variables')
plt.ylabel('Values')

# Show the plot
plt.show()
```

### Sub Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file (adjust file path as necessary)
file_path = '10years GannavaramData.csv'
data = pd.read_csv(file_path)

# Print the columns to understand their structure
print(data.columns)

# Rename columns based on the actual structure
data.columns = ["Date", "Station", "Column3", "Max Temp", "Column5", "Min Temp", "Column7", "RH% 830", "RH% 1730", "Column10", "Rainfall"]

# Drop unnecessary columns (Column3, Column5, Column7, Column10)
data = data.drop(columns=["Column3", "Column5", "Column7", "Column10"])

# Convert relevant columns to numeric values
data["Max Temp"] = pd.to_numeric(data["Max Temp"], errors='coerce')
data["Min Temp"] = pd.to_numeric(data["Min Temp"], errors='coerce')
data["Rainfall"] = pd.to_numeric(data["Rainfall"], errors='coerce')

# Remove any rows where 'Date' is not numeric
data = data[data["Date"] != "Date"]
data["Date"] = pd.to_numeric(data["Date"], errors='coerce')

# Prepare the data for the violin plot
melted_data = data.melt(value_vars=["Max Temp", "Min Temp", "Rainfall"], 
                         var_name="Variable", value_name="Value")

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Weather Data Analysis')

# Box plot for Max Temp
sns.boxplot(ax=axs[0, 0], x='Variable', y='Value', data=melted_data[melted_data['Variable'] == 'Max Temp'],color='dodgerblue')
axs[0, 0].set_title('Box Plot of Max Temp')
axs[0, 0].set_ylabel('Max Temperature')

# Box plot for Min Temp
sns.boxplot(ax=axs[0, 1], x='Variable', y='Value', data=melted_data[melted_data['Variable'] == 'Min Temp'],color='forestgreen')
axs[0, 1].set_title('Box Plot of Min Temp')
axs[0, 1].set_ylabel('Min Temperature')

# Violin plot for Rainfall
sns.violinplot(ax=axs[1, 0], x='Variable', y='Value', data=melted_data[melted_data['Variable'] == 'Rainfall'], inner='quartile',hue='Variable',palette='pastel')
axs[1, 0].set_title('Violin Plot of Rainfall')
axs[1, 0].set_ylabel('Rainfall')

# Remove the empty subplot (2,1)
axs[1, 1].axis('off')  # This leaves the last subplot empty

# Adjust layout
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
